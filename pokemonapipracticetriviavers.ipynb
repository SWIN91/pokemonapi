```python
def new_game():
    guesses = []          #using this to keep track of user input
    correct_guesses = 0   #using this to keep track of points
    question_num = 1     #using this variable as some sort of counter...see line 9
    
    for key in questions:
        print("-------------------------------------------------------------------------------------------")
        print(key)
        for i in pokemon[question_num-1]: #subtract one to receive 0 as the index and move to the next question
            print(i)
        guess = input("Enter (A, or B): ")
        guess = guess.upper() #remedy the case-sensitivity of strings
        guesses.append(guess)
        correct_guesses += check_answer(questions.get(key), guess) #key = answer, guess = guess, increment crrct guesses
        question_num += 1      #increment question number after each iteration    
    display_score(correct_guesses, guesses)    
def check_answer(answer, guess):
    if answer == guess:
        print("CORRECT!")
        return 1
    else:
        print("INCORRECT!")
        return 0
def display_score(correct_guesses, guesses):
    print("-------------------------------------------------------------------------------------------")
    print("RESULTS")
    print("-------------------------------------------------------------------------------------------")
    print("Answers: ", end=" ") #so it doesn't start on a new line
    for i in questions:
        print(questions.get(i), end=" ")
    print()
    print("Guesses: ", end=" ")
    for i in guesses:
        print(i, end=" ")
    print()
    score = (correct_guesses/len(questions))*100
    print("YOUR SCORE:" + str(score) +"%") # have to typecast int to a string
def play_again():
    response = input("Do you want to play again? (Yes or No): ")
    response = response.upper()
    if response == "YES":
        return True
    else:
        return False

questions = {
"Who is the god of all Pokemon?: " : "A",
"Which Legendary Pokemon did Ash see on the first day of his Pokemon journey?: " : "A",
"Which Pokemon is the oldest in the game universe?: " : "B",
"Which was the first Pokemon Ash ever caught?: " : "A",
"Which Pokemon is able to control time? " : "B",
"Which Pokemon is always holding a pendelum? " : "A",
"What does Puputar evolve into? " : "B",
"Which Pokemon can be resurrected from the old amber? " : "B",
"Which Pokemon can change its gender? " : "A",
"Which Pokemon can be resurrected from the Helix Fossil? " : "A"
}

pokemon = [["A. Arceus", "B. Aerodactyl"],         #change my pokemon to my options
           ["A. Ho-oh", "B. Dialga"],
           ["A. Aerodactyl", "B. Arceus"],
           ["A. Caterpie", "B. Eevee"],
           ["A. Hypno", "B. Dialga"],
           ["A. Hypno", "B. Psyduck"],
           ["A. Flygon", "B. Tyranitar"],
           ["A. Omanyte", "B. Aerodactyl"],
           ["A. Azurill", "B. Ditto"],
           ["A. Omanyte", "B. Aerodactyl"]
          ]


new_game()
while play_again():
    new_game()
print("Bye!")
    
```
